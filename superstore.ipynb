```python
# find out different shipment mode and the respective percentages.
k = pd.DataFrame(order["Ship Mode"])
Reg_Air = (len(k[k["Ship Mode"] == "Regular Air"])/len(k))*100
Express_Air = (len(k[k["Ship Mode"] == "Express Air"])/len(k))*100
Delivery_Truck = (len(k[k["Ship Mode"] == "Delivery Truck"])/len(k))*100
print("Mode :Regular Air Percentage =",Reg_Air,"\n","Mode :Express Air Percentage =",Express_Air,"\n","Mode :Delivery Truck")
```

```python
# Create a new column and try to find out difference between order date and shipment date
order["Delivery_Time(Days)"] = order["Ship Date"] - order["Order Date"]
order["Delivery_Time(Days)"]
```

```python
# for que 8 find out where duration is more than 10 days.

#Converting timedelta format to str/object
order["Delivery_Time(Days)"] = order["Delivery_Time(Days)"].astype(str)

#Strip "days" from "Delivery_Time(Days)" column
order["Delivery_Time(Days)"] = order["Delivery_Time(Days)"].str.replace("days","")

#Converting to int format to perform operations
order["Delivery_Time(Days)"] = order["Delivery_Time(Days)"].astype(int)

order[order["Delivery_Time(Days)"] > 10]
```

```python
# find returned shipments for which number of days >15

#inner join to find common orders
return_orders = pd.merge(order,ret,how = "inner")

#finding number of days
return_orders["Delivery_Time(Days)"] = order["Ship Date"] - order["Order Date"]

#converting column to str/object type
return_orders["Delivery_Time(Days)"] = return_orders["Delivery_Time(Days)"].astype(str)

#removing the substring "day"
return_orders["Delivery_Time(Days)"] = return_orders["Delivery_Time(Days)"].str.replace("days","")

#Converting to int to perform operations
return_orders["Delivery_Time(Days)"] = return_orders["Delivery_Time(Days)"].astype(int)

return_orders[return_orders["Delivery_Time(Days)"] > 15 ]
```

```python
# group by region and find out the most profitable region.
max_profit_region = order[["Region","Profit"]].groupby("Region").sum().reset_index()
max_profit_region[max_profit_region["Profit"] == max(max_profit_region["Profit"])]
```

```python
# try to find overall in which country/region more discount is offered.
max_disc_region = order[["Region","Discount"]].groupby("Region").sum().reset_index()
max_disc_region[max_disc_region["Discount"] == max(max_disc_region["Discount"])]

```
