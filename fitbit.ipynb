```python
# how many of them have not logged there activity find out in terms of number of ids
df2 = pd.DataFrame(df[df['LoggedActivitiesDistance'] == 0]['Id'].unique(), columns=['ID'])
print('{}  IDs have not logged their activity'.format(len(df2)))
```

```python
# Explore over an internet that how much calories burn is required for a healthy person and find out how many healthy person we have in our dataset
df['Daily_req'] = df['Calories'].apply(lambda x: 'Y' if x > 1500 else 'N')
df3 = pd.DataFrame(df.groupby('Id')['Daily_req'].count())
df3['cnt_Y'] = pd.DataFrame(df[df['Daily_req'] == 'Y'].groupby('Id')['Daily_req'].count())
# A person is healty if he atleast burns 1500 calories a day for 3/4th of the given record.
df3['%_active_days'] = round(100 * df3['cnt_Y'] / df3['Daily_req'], 0)
df3['is Healthy'] = df3['%_active_days'].apply(lambda x: 'Y' if x > 75 else 'N')
df3[df3['is Healthy'] == 'Y']
print('{} no of healthy person'.format(len(df3[df3['is Healthy'] == 'Y'])))
```

```python
# how many person are not a regular person with respect to activity try to find out those

df4 = df3[df3['%_active_days'] < 50]
df4
```

```python
#who is the thired most active person in this dataset find out those
df5 = pd.DataFrame(df.groupby(['Id'])['Calories'].mean()).reset_index()
df5['Rank'] = df5['Calories'].rank(method='dense', ascending=False).astype('int')
df5[df5['Rank'] == 3]
```

```python
# who is the 5th most laziest person avilable in dataset find it out
df5['Lazy_rnk'] = df5['Calories'].rank(method='dense').astype('int')
df5[df5['Lazy_rnk'] == 5]
```
